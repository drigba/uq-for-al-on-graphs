```python
def get_count_dict(t):
    ixs = torch.tensor([l[1:] for l in t["acquired_idxs"]]).flatten()
    count = torch.bincount(ixs)
    keys = torch.where(count)
    count_dict = {k.item():count[k].item() for k in keys[0]}
    return count_dict, count,ixs
```

```python
count_dict_sgc,count_sgc, ixs_sgc = get_count_dict(t_geem)
print(f"Maximally acquired index: {count_sgc.argmax()} with {count_sgc.max()} acquisitions")

count_dict_gcn,count_gcn, ixs_gcn = get_count_dict(t_gcn_alea_tta_filter_prob)
print(f"Maximally acquired index: {count_gcn.argmax()} with {count_gcn.max()} acquisitions")
print("------")

print(f"Intersection of acquired indices: {len(set(count_dict_sgc.keys()).intersection(set(count_dict_gcn.keys())))}")
print(f"Union of acquired indices: {len(set(count_dict_sgc.keys()).union(set(count_dict_gcn.keys())))}")
print(f"SGC acquired indices: {len(count_dict_sgc)}")
print(f"GCN acquired indices: {len(count_dict_gcn)}")
```
